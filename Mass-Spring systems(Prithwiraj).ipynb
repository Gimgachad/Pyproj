```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# parameters
m = 1.0  # mass
k = 1.0  # spring constant

# system of differential equations
def system(y, t):
    x, v = y
    dx = v
    dv = -k*x / m
    return [dx, dv]

# initial conditions
x0 = 0.0  # initial position
v0 = 1.0  # initial velocity
y0 = [x0, v0]

# time grid
t = np.linspace(0, 40, 1000)

# solve the ODE
sol = odeint(system, y0, t)

# plot the results
plt.figure(figsize=(14,6))
plt.grid(True)
plt.plot(sol[:, 0], sol[:, 1])
plt.title("Normal mass-spring system")
plt.xlabel('velocity')
plt.ylabel('Displacement')
plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define the symbols
t = sp.symbols('t')

# Parameters
m_val = 1  # Example value for mass
k_val = 1  # Example value for spring constant

# Differential equation
x = sp.Function('x')(t)
eq = sp.Eq(m_val * x.diff(t, 2) + k_val * x, 0)

# Solve the differential equation
sol = sp.dsolve(eq)

# Print the solution
display(eq)
display(sol)

# Define the solution function manually
def x_func(t_val):
    C1, C2 = sp.symbols('C1 C2')
    x_sol = sol.rhs.subs({'C1': 1, 'C2': 0})  # Assuming initial conditions x(0) = 1 and x'(0) = 0
    return x_sol.subs(t, t_val).evalf()

# Generate time values for plotting
t_values = np.linspace(0, 40, 1000)

# Evaluate the solution function at the given time values
x_values = [x_func(t_val) for t_val in t_values]

# Plot the graph
plt.figure(figsize=(14, 6))
plt.plot(t_values, x_values, label='Solution')
plt.xlabel('Time')
plt.ylabel('Position')
plt.title('Solution of the Harmonic Oscillator Equation')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# parameters
m = 1.0  # mass
k = 1.0  # spring constant
b = 0.2  # damping coefficient
F0 = 1.0  # amplitude of external force
omega = 3  # frequency of external force

# system of differential equations
def system(y, t):
    x, v = y
    dx = v
    dv = (F0*np.sin(omega*t) - b*v - k*x) / m
    return [dx, dv]

# initial conditions
x0 = 0.0  # initial position
v0 = 0.0  # initial velocity
y0 = [x0, v0]

# time grid
t = np.arange(0, 80, 0.0001)

# solve the ODE
sol = odeint(system, y0, t)

# plot the results
plt.figure(figsize=(14,6))
plt.plot(sol[:, 0], sol[:, 1])
plt.grid(True)
plt.title("Forced mass-spring system")
plt.xlabel('Velocity')  
plt.ylabel('Displacement')

plt.show()
```
