```python
#Plot of Gaussian wave packet
import numpy as np
import matplotlib.pyplot as plt
def gaussian_wave_function(x, x0, sigma, k):
    psi = np.exp(-(x - x0)**2 / (2 * sigma**2)) * np.exp(1j * k * x)
    return psi
# Parameters
x = np.linspace(-10, 10, 1000) # x-axis values
x0 = 0 # Average position
sigma = 1 # Width of the wave function
k = 1 # Wave number
# Compute the wave function
wave_function = gaussian_wave_function(x, x0, sigma, k)
# Plotting the wave function
#plt.plot(x, wave_function.real, label='Real part')
#plt.plot(x, wave_function.imag, label='Imaginary part')
plt.plot(x, np.abs(wave_function)**2, label='Probability density')
# Customize the plot
plt.title('Gaussian Wave Function')
plt.xlabel('Position (x)')
plt.ylabel('Wave function')
plt.legend()
# Display the plot
plt.show()
```

```python
# Plot of Lorentzian wavepacket.
import numpy as np
import matplotlib.pyplot as plt
def lorentzian_wave_packet(x, x0, gamma, k0):
    psi = (1 / (np.pi * gamma)) * (gamma**2 / ((x - x0)**2 + gamma**2)) * np.exp(1j * k0 * x)
    return psi
# Parameters
x = np.linspace(-10, 10, 1000) # x-axis values
x0 = 0 # Average position
gamma = 1 # Width parameter (also known as the half-width at half-maximum)
k0 = 1 # Average momentum
# Compute the wave function
wave_packet = lorentzian_wave_packet(x, x0, gamma, k0)
# Plotting the wave packet
#plt.plot(x, wave_packet.real, label='Real part')
#plt.plot(x, wave_packet.imag, label='Imaginary part')
plt.plot(x, np.abs(wave_packet)**2, label='Probability density')
# Customize the plot
plt.title('Lorentzian Wave Packet')
plt.xlabel('Position (x)')
plt.ylabel('Wave function')
plt.legend()
# Display the plot
plt.show()
```

```python
 # Gauss Distribution function
import numpy as np
import matplotlib.pyplot as plt
def gaussian_distribution(x, mu, sigma):
    coeff = 1 / (sigma * np.sqrt(2 * np.pi))
    exponent = -0.5 * ((x - mu) / sigma)**2
    return coeff * np.exp(exponent)
# Parameters
x = np.linspace(-10, 10, 1000) # x-axis values
mu = 0 # Mean of the Gaussian distribution
sigma = 1 # Standard deviation of the Gaussian distribution
# Compute the distribution values
distribution = gaussian_distribution(x, mu, sigma)
# Plotting the Gaussian distribution
plt.plot(x, distribution)
# Customize the plot
plt.title('Gaussian Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
# Display the plot
plt.show()
```

```python
# Plot of Planck Distribution law
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import h, c, k
# Planck's Distribution Law
def plancks_law(wavelength, temperature):
    x = (h * c) / (wavelength * k * temperature)
    return (8 * np.pi * h * c**2) / (wavelength**5 * np.expm1(x))
# Parameters
temperature = 5000 # Temperature of the black body in Kelvin
# Wavelength range (in meters)
wavelengths = np.linspace(1e-9, 3e-6, 1000)
# Calculate intensities using Planck's law
intensities = plancks_law(wavelengths, temperature)
# Plotting Planck's Distribution Law
plt.plot(wavelengths * 1e9, intensities, label=f'Temperature = {temperature} K')
# Customize the plot
plt.title("Planck's Distribution Law")
plt.xlabel('Wavelength (nm)')
plt.ylabel('Intensity')
plt.legend()
plt.grid(True)
# Display the plot
plt.show()
```

```python
# Maxwell velocity Distribution law
def maxwell_boltzmann_speed_distribution(v, T, m):
    k = 1.380649e-23 # Boltzmann constant in J/K
    coeff = (m / (2 * np.pi * k * T)) ** (3 / 2)
    exponent = - (m * v ** 2) / (2 * k * T)
    return coeff * np.exp(exponent)
# Parameters
v = np.linspace(0, 5000, 1000) # Velocity values in m/s
T = 300 # Temperature in Kelvin
m = 1.67e-27 # Particle mass in kg (e.g., for an ideal gas particle)
# Compute the distribution values
distribution = maxwell_boltzmann_speed_distribution(v, T, m)
# Plotting the Maxwell-Boltzmann velocity distribution
plt.plot(v, distribution)
# Customize the plot
plt.title('Maxwell-Boltzmann Velocity Distribution')
plt.xlabel('Velocity (m/s)')
plt.ylabel('Probability Density')
# Display the plot
plt.show()
```

```python
# Constants
h = 6.62607015e-34 # Planck's constant in J*s
c = 2.99792458e8 # Speed of light in m/s
k = 1.380649e-23 # Boltzmann constant in J/K
# Frequency range
frequency_range = np.linspace(1e12, 3e15, 1000) # Frequency range in Hz
# Function to calculate spectral radiation density
def spectral_radiation_density(temperature, frequency):
    energy = h * frequency
    intensity = (2 * h * frequency**3 / c**2) * (1  / (np.exp(energy / (k * temperature)) - 1))
    return intensity
# Temperature values
temperature_values = [300, 1000, 5000, 10000] # Temperature values in Kelvin
# Plotting the spectral radiation density at different temperatures
plt.figure(figsize=(10, 6))
for temperature in temperature_values:
    intensity_data = spectral_radiation_density(temperature, frequency_range)
    plt.plot(frequency_range / 1e12, intensity_data, label=f'Temperature = {temperature} K')
plt.xlabel('Frequency (THz)')
plt.ylabel('Spectral Radiation Density (W/(m^2*Hz*sr))')
plt.title('Spectral Radiation Density vs. Frequency at Different Temperatures')
plt.legend()
plt.show()
```
