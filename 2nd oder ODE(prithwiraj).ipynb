```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
 
 
def rhsODE(x,t,k1,k2):
    dxdt = [x[1], -k1 * x[1]-k2 * np.sin(x[0])]
    return dxdt
 
# set the initial conditions
x0 = [0, 1]
 
# define the discretization points
t_span = np.linspace(0, 100, 300)
 
#define the constants 
k1 = 0.1
k2 = 0.5
 
solutionOde = odeint(rhsODE, x0, t_span, args=(k1,k2))

plt.figure(figsize=(14, 7))
plt.plot(t_span, solutionOde[:, 0], 'b', label='x1')
plt.plot(t_span, solutionOde[:, 1], 'g', label='x2')
plt.legend(loc='best')
plt.xlabel('time')
plt.ylabel('x1(t), x2(t)')
plt.grid()
plt.savefig('simulation.png')
plt.show()
```
