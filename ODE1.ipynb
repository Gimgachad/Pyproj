```python
import numpy as np
import matplotlib.pyplot as plt

# Define the differential equation
def ode_4_c(t, x):
    return 3 * x**2 - 4 * x

# Define the Runge-Kutta 4th order solver
def solve_rk4(ode, t, y0):
    y0 = np.array(y0)
    y = np.zeros([len(t), len(y0)])  # Initialize the array with len(t) rows and len(y0) columns
    y[0] = y0
    h = (t[-1] - t[0]) / (len(t) - 1)
    
    for i in range(1, len(t)):
        k1 = h * ode(t[i-1], y[i-1])
        k2 = h * ode(t[i-1] + h/2, y[i-1] + k1/2)
        k3 = h * ode(t[i-1] + h/2, y[i-1] + k2/2)
        k4 = h * ode(t[i], y[i-1] + k3)
        
        y[i] = y[i-1] + (k1 + 2*k2 + 2*k3 + k4) / 6
    
    return y

# Define the time array with more points for a smoother graph
t = np.linspace(1, 1.1, 10000)  # Adjust the range to prevent overflow
x1 = [5]

# Solve the ODE using the RK4 solver
sol = solve_rk4(ode_4_c, t, x1)

# Plot the solution
plt.figure(figsize=(12, 6))
plt.plot(t, sol)
plt.xlabel('t')
plt.ylabel('x')
plt.grid(True)
plt.title("Graph of the ODE using RK4")
plt.show()
```

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define the ODE system for solve_ivp
def ode_system(t, y):
    x, x_prime = y
    dx_dt = x_prime
    dx_prime_dt = -2 * t**2 * x_prime - x
    return [dx_dt, dx_prime_dt]

# Initial conditions and time points
x0 = [1, 0]
t_span = (0, 10)
t_eval = np.linspace(0, 10, 1000)

# Solve using solve_ivp with high tolerance for accuracy
sol_ivp = solve_ivp(ode_system, t_span, x0, t_eval=t_eval, method='RK45', rtol=1e-10, atol=1e-10)

# Solve using sympy
t = sp.symbols('t')
x = sp.Function('x')(t)
eq = sp.Eq(x.diff(t, 2) + 2 * t**2 * x.diff(t, 1) + x, 0)
sol = sp.dsolve(eq, ics={x.subs(t, 0): 1, x.diff(t).subs(t, 0): 0})

# Manually remove the O term by using series expansion and discarding the big-O notation
sol_series = sol.rhs.series(t, 0, 6).removeO()

# Convert the simplified solution to a numerical function
sol_func = sp.lambdify(t, sol_series, 'numpy')

# Evaluate the symbolic solution at the numerical points
x_list = np.array([sol_func(i) for i in t_eval], dtype=np.float64)

# Plotting the solutions
plt.figure(figsize=(14, 8))
plt.plot(sol_ivp.t, sol_ivp.y[0], label='solve_ivp', color='blue')
plt.plot(t_eval, x_list, '--', label='sympy', color='orange')
plt.xlabel('t')
plt.ylabel('x')
plt.legend()
plt.title("Comparison of Solution of the ODE using solve_ivp and SYMPY (Analytical)")
plt.grid(True)
plt.show()
```

```python
from ode_casilab import solve_rk4
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Define parameters
m1, m2 = 1, 1  # masses
k1, k2, k3 = 1, 1, 1  # spring constants

# Define the ODE system
def ode_system(t, y):
    x1, v1, x2, v2 = y
    dx1dt = v1
    dv1dt = (-k1 * x1 + k2 * (x2 - x1)) / m1
    dx2dt = v2
    dv2dt = (-k3 * x2 - k2 * (x2 - x1)) / m2
    return [dx1dt, dv1dt, dx2dt, dv2dt]

# Define the time points
t = np.linspace(0, 20, 1000)
# Initial conditions: [x1(0), v1(0), x2(0), v2(0)]
y0 = [1, 0, 0, 0]

# Solve using solve_rk4
sol_rk4 = solve_rk4(ode_system, t, y0)
x1_sol_rk4 = sol_rk4[:, 0]
v1_sol_rk4 = sol_rk4[:, 1]
x2_sol_rk4 = sol_rk4[:, 2]
v2_sol_rk4 = sol_rk4[:, 3]

# Solve using SymPy
t_sym = sp.symbols('t')
x1_sym, x2_sym = sp.symbols('x1 x2', cls=sp.Function)
v1_sym, v2_sym = x1_sym(t_sym).diff(t_sym), x2_sym(t_sym).diff(t_sym)

eq1 = m1 * x1_sym(t_sym).diff(t_sym, t_sym) + k1 * x1_sym(t_sym) - k2 * (x2_sym(t_sym) - x1_sym(t_sym))
eq2 = m2 * x2_sym(t_sym).diff(t_sym, t_sym) + k3 * x2_sym(t_sym) + k2 * (x2_sym(t_sym) - x1_sym(t_sym))

sol_sym = sp.dsolve((eq1, eq2), ics={x1_sym(0): y0[0], x2_sym(0): y0[2], x1_sym(t_sym).diff(t_sym).subs(t_sym, 0): y0[1], x2_sym(t_sym).diff(t_sym).subs(t_sym, 0): y0[3]})
x1_sol_sym = sol_sym[0].rhs
x2_sol_sym = sol_sym[1].rhs

# Convert the SymPy solutions to numerical functions
x1_sol_func = sp.lambdify(t_sym, x1_sol_sym, 'numpy')
x2_sol_func = sp.lambdify(t_sym, x2_sol_sym, 'numpy')

# Evaluate the solutions
x1_sol_vals_sym = x1_sol_func(t)
x2_sol_vals_sym = x2_sol_func(t)
v1_sol_vals_sym = np.gradient(x1_sol_vals_sym, t)
v2_sol_vals_sym = np.gradient(x2_sol_vals_sym, t)

# Plot the time-trajectory comparison
plt.figure(figsize=(14, 8))
plt.plot(t, x1_sol_rk4, label='x1 (RK4)')
plt.plot(t, x2_sol_rk4, label='x2 (RK4)')
plt.plot(t, x1_sol_vals_sym, '--', label='x1 (SymPy)')
plt.plot(t, x2_sol_vals_sym, '--', label='x2 (SymPy)')
plt.xlabel('Time')
plt.ylabel('Displacement')
plt.title('Time-Trajectory Comparison')
plt.legend()
plt.grid()
plt.show()

# Plot the phase-space diagram comparison
plt.figure(figsize=(14, 8))
plt.plot(x1_sol_rk4, v1_sol_rk4, label='Phase Space x1-v1 (RK4)')
plt.plot(x2_sol_rk4, v2_sol_rk4, label='Phase Space x2-v2 (RK4)')
plt.plot(x1_sol_vals_sym, v1_sol_vals_sym, '--', label='Phase Space x1-v1 (SymPy)')
plt.plot(x2_sol_vals_sym, v2_sol_vals_sym, '--', label='Phase Space x2-v2 (SymPy)')
plt.xlabel('Displacement')
plt.ylabel('Velocity')
plt.title('Phase-Space Diagram Comparison')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

def solve_rk4(ode, t, y0):
    y0 = np.array(y0)
    y = np.zeros((len(t), len(y0)))
    y[0] = y0
    h = t[1] - t[0]
    for i in range(1, len(t)):
        k1 = h * np.array(ode(t[i-1], y[i-1]))
        k2 = h * np.array(ode(t[i-1] + h/2, y[i-1] + k1/2))
        k3 = h * np.array(ode(t[i-1] + h/2, y[i-1] + k2/2))
        k4 = h * np.array(ode(t[i-1] + h, y[i-1] + k3))
        y[i] = y[i-1] + (k1 + 2*k2 + 2*k3 + k4) / 6
    return y

def projectile(t, y):
    y, y1 = y
    dy_dt = y1
    dy1_dt = -g
    return np.array([dy_dt, dy1_dt])

# Parameters
t = np.linspace(0, 1, 1000)  # Time range of 1 second
y0 = [0, 1]  # Initial position and velocity
g = 9.8  # Acceleration due to gravity

# Solve using solve_rk4
sol_rk4 = solve_rk4(projectile, t, y0)

# Solve using SymPy
t_sym = sp.Symbol('t')
y_sym = sp.Function('y')(t_sym)

# Define the ODE
ode_sym = sp.Eq(y_sym.diff(t_sym, t_sym), -g)

# Display the ODE
print("Differential equation:")
display(ode_sym)

# Use the known solution for projectile motion
y_sol_expr = y0[1] * t_sym - 0.5 * g * t_sym**2 + y0[0]

# Display the solution
print("\nSolution:")
display(sp.Eq(y_sym, y_sol_expr))

# Convert the solution to a numpy function
y_sol_func = sp.lambdify(t_sym, y_sol_expr, 'numpy')

# Evaluate the SymPy solution
y_sol_sym = y_sol_func(t)
v_sol_sym = y0[1] - g * t  # Analytical solution for velocity

# Plot the time trajectory
plt.figure(figsize=(10, 6))
plt.plot(t, sol_rk4[:, 0], label='RK4')
plt.plot(t, y_sol_sym, '--', label='Analytical')
plt.title('Time Trajectory Diagram')
plt.xlabel('Time (s)')
plt.ylabel('Height (m)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot the phase space
plt.figure(figsize=(10, 6))
plt.plot(sol_rk4[:, 0], sol_rk4[:, 1], label='RK4')
plt.plot(y_sol_sym, v_sol_sym, '--', label='Analytical')
plt.title('Phase Space Diagram')
plt.xlabel('Position (m)')
plt.ylabel('Velocity (m/s)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from ode_casilab import solve_rk4
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from scipy.special import airy

# Define symbolic variables
x_sym = sp.Symbol('x')
y = sp.Function('y')(x_sym)

# Define the second-order ODE: y'' = x * y
second_order_ode = sp.Eq(y.diff(x_sym, x_sym), x_sym * y)

# Solve the second-order ODE symbolically with initial conditions
solution = sp.dsolve(second_order_ode, y, ics={y.subs(x_sym, 0): 1, y.diff(x_sym).subs(x_sym, 0): 2})

# Extract the solution
y_sol_sym = solution.rhs

# Compute the Airy function solution numerically
def compute_airys(x):
    # Airy functions Ai and Bi
    airy_sol = airy(x)
    return airy_sol[0], airy_sol[1]  # Ai(x), Bi(x)

# Numeric setup
x = np.linspace(0, 10, 100)
y0 = [1, 2]

# Define the system for numerical solving
def airy_system(x, y):
    y, y1 = y
    dy_dx = y1
    dy1_dx = x * y
    return np.array([dy_dx, dy1_dx])

# Solve the ODE numerically using RK4
sol_numeric = solve_rk4(airy_system, x, y0)

# Compute the analytical solutions
y_sol_analytical, y1_sol_analytical = compute_airys(x)

plt.figure(figsize=(12, 6))

# Time trajectory diagram
plt.subplot(121)
plt.plot(x, sol_numeric[:, 0], label='Numerical y')
plt.plot(x, y_sol_analytical, label='Analytical y', linestyle='--')
plt.title('Time Trajectory Diagram')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Phase space diagram
plt.subplot(122)
plt.plot(sol_numeric[:, 0], sol_numeric[:, 1], label='Numerical Phase Space')
plt.plot(y_sol_analytical, y1_sol_analytical, label='Analytical Phase Space', linestyle='--')
plt.title('Phase Space Diagram')
plt.xlabel('y')
plt.ylabel('dy_dx')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
