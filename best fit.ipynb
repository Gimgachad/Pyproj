```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Sample CSV data (replace with your actual data)
data = """
Voltage, Current
0.2,     1.3
0.4,     2.7
0.6,     4
0.8,     5.3
1,       6.7
1.2,     8
1.4,     9.3
1.6,     10.7
1.8,     12
2,       13.3
2.2,     14.16
2.4,     18
2.6,     17.3
2.8,     18.7
3,       20    
3.2,     21.3
3.4,     22.7
3.6,     24
3.8,     25.3
4,       26.7
"""

# Read data from CSV string
lines = data.strip().split('\n')
headers = lines[0].split(', ')
data = np.array([row.split(',') for row in lines[1:]], dtype=float)
Voltage = data[:, 0]
Current = data[:, 1]

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(Voltage, Current)

# Calculate the best fit line
best_fit_line = slope * Voltage + intercept

# Plot the data and the best fit line
plt.scatter(Voltage, Current, label='Data')
plt.plot(Voltage, best_fit_line, color='red', label='Best Fit Line')

# Add labels and title
plt.xlabel(headers[0])
plt.ylabel(headers[1])
plt.title('Best Fit Line for ' + headers[0] + 'vs. '+ headers[1])

# Add legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
