### Newton-Gregory forward differentiation

```python
# Code for Newton's Forward Differentiation Method (FDM)
import math
x=0.1
h=0.0001
f = lambda x: 7*x ** 3-4*x ** 2+8*x+10
devt1=(f(x+h)-f(x))/h
devt2=(f(x+2*h)-2*f(x+h)+f(x))/h ** 2
print(f"First derivative of f(x) using FDM is:" , (devt1))
print(f"Second derivative of f(x) using FDM is:", (devt2))
```

### Newton-Gregory central differentiation

```python
# Code for Newton's Central Differentiation Method (CDM)
import math
x=0.1
h=0.0001
f = lambda x: 7*x ** 3-4*x ** 2+8*x+10
devt1=(f(x+h/2)-f(x-h/2))/h
devt2=(f(x+h)-2*f(x)+f(x-h))/h ** 2
print(f"First derivative of f(x) using CDM is:" , (devt1))
print(f"Second derivative of f(x) using CDM is:", (devt2))
```

```python
#Integration using built-in Trapezoid Module
import numpy as np
from scipy.integrate import trapz

a = 0
b = np.pi/2
n = 11
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = (np.sin(x))*(np.cos(x))
result_trap = trapz(f,x)
error_trap = (h/2)*(f[0] + 2 * sum(f[1:n-1]) + f[n-1])
print("Integration output using built-in Trapezoidal module:",
result_trap)
print("Error using built-in Trapezoidal module:", error_trap)
```

```python
# Method I: Code for numerical integration using Trapezoid Rule UDEMY)
import numpy as np
from math import *
import matplotlib.pyplot as plt

a = 0;
b = np.pi/2;
n = 1000
h = (b-a)/n
f = lambda x: (np.sin(x))*(np.cos(x))

S = 0.5*(f(a)+f(b))
for i in range(1,n-1):
    S += f (a+i*h)
Integral = h*S
print ("Integration output using Trapezoidal module:", Integral)
```

```python
# Udemy (SIMPSON 'S 1/3 RULE) 
import numpy as np 
f = lambda x: (np.sin(x) ) *(np.cos(x)) #math. sin(x) 
a = 0 
b = np.pi/2 
n = 1000 
h= (b - a)/n 
S= (f(a) +f (b)) 
for i in range (1,n-1,2): 
    S+=4*f(a+i*h) 
for i in range (2,n-2,2): 
    S+=2*f(a+i*h) 
Integral=(h/3) *S 
print("Integration output using Simpson 1/3 Rule:",    Integral)
```

```python
#Python Programming And Numerical Methods : A Guide For Engineers And Scientists (SIMPSON'S 1/3 RULE) (p. 362) 
import numpy as np 
b= np.pi/2 
n = 101 
h = (b - a) / (n - 1) 
x = np.linspace(a, b, n) 
f = (np.sin(x))*(np. cos (x)) 
result_sim = (h/3) * (f[0] + 2*sum(f[0:n-2:2]) + 4*sum (f [1:n-1:2]) + f[n-1] ) 
error_simp = 2 - result_sim 
print("Integration output using Simpson Rule:", result_sim)
print("Error using simpson Rule:", error_simp)
```

```python
# Udemy (SIMPSON'S 3/8 RULE)
import math
import numpy as np
f = lambda x: (np.sin(x))*(np.cos(x)) #math.sin(x)
a = 0
b= np.pi/2
n = 100
h = (b - a) / n
S = f(a) +f (b)
for i in range(1,n,3):
    S+=3*(f(a+i*h)+f(a+(i+1)*h))
for i in range(3,n,3):
    S+=2*f(a+i*h)
Integral=(3*h/8)*S
print("Integration output using Simpson 3/8 Rule:", Integral)
```
