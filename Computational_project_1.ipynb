```python
# (i) Lotka-Voltera equations
# x'(t) = (alpha)x - (beta)xy ; y'(t) = (delta)xy - gamma

from ode_casilab import solve_rk4
import numpy as np 
import matplotlib.pyplot as plt 

def LVmodel(t, x_y):
    alpha, beta, gamma, delta = 1, 0.1, 1, 0.1 # # Constants
    x,y = x_y
    dx_dt = alpha*x - beta*x*y
    dy_dt = delta*x*y - gamma*y
    return np.array([dx_dt, dy_dt])

t = np.linspace(0,20,1000)
x0_y0 = [1,1]

sol = solve_rk4(LVmodel, t, x0_y0)

plt.plot(t, sol[:,0],label='x')
plt.plot(t, sol[:,1],label='y')
plt.xlabel('t')
plt.ylabel('x, y')
plt.legend()
plt.show()
```

```python
# (e) Forced-Damped SHO y" + gamma y' + w^2y = F_0 cos(wt) 
# Let y' = y1 so that y1' + (gamma)y1 + w^2y = F_0 cos(wt)
# (i) Over damped system (gamma > w) 
# (ii) Under damped syatem (gamma < w) 
# (iii) Critically damped system (gamma = w)

from ode_casilab import solve_rk4
import numpy as np 
import matplotlib.pyplot as plt 

def forced_shm(t, y):
    y, y1 = y
    w = 1
    F_0 = 1
    
    dy_dt = y1 
    dy1_dt = -gamma*y1 - w**2 * y + F_0*np.cos(w*t)
    return np.array([dy_dt, dy1_dt])

t = np.linspace(0,20,1000)
y0 = [1,0]

plt.figure(figsize=(12,16))

# (i) Over damped system (gamma > w) 
plt.subplot(321)
gamma = 2
sol = solve_rk4(forced_shm, t, y0)
plt.plot(t, sol[:,0],label='y')
#plt.plot(t, sol[:,1],label='dy_dt')
plt.title('Time trajectory diagram (gamma > w)')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.subplot(322)
plt.plot(sol[:,0], sol[:,1])
plt.title('Phase space diagram(gamma > w)')
plt.xlabel('y')
plt.ylabel('dy_dt')
plt.grid(True)

# (ii) Under damped syatem (gamma < w) 
plt.subplot(323)
gamma = .5
sol = solve_rk4(forced_shm, t, y0)
plt.plot(t, sol[:,0],label='y')
#plt.plot(t, sol[:,1],label='dy_dt')
plt.title('Time trajectory diagram (gamma < w) ')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.subplot(324)
plt.plot(sol[:,0], sol[:,1])
plt.title('Phase space diagram (gamma < w)')
plt.xlabel('y')
plt.ylabel('dy_dt')
plt.grid(True)

# (iii) Critically damped system (gamma = w)
plt.subplot(325)
gamma = 1
sol = solve_rk4(forced_shm, t, y0)
plt.plot(t, sol[:,0],label='y')
#plt.plot(t, sol[:,1],label='dy_dt')
plt.title('Time trajectory diagram (gamma = w)')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.subplot(326)
plt.plot(sol[:,0], sol[:,1])
plt.title('Phase space diagram (gamma = w)')
plt.xlabel('y')
plt.ylabel('dy_dt')
plt.grid(True)

plt.show()
```

```python
# (f) AC driven LCR circuit: Lq"(t) + Rq'(t) + q(t)/C = q_0*sin(wt)
#  L = 1H, C = 0.01F, R = 25ohm, q_0 = 1C, w = 100pi Hz
# Let q' = I so that LI'(t) + RI(t) + q(t)/C = q_0*sin(wt)
# or I'(t) + 25I(t) + q(t)/0.01 = sin(100pi t)

from ode_casilab import solve_rk4
import numpy as np 
import matplotlib.pyplot as plt 

def LCR(t, q):
    q, I = q
    dq_dt = I 
    dI_dt = -25*I - q/0.01 + np.sin(100*np.pi*t)
    return np.array([dq_dt, dI_dt])

t = np.linspace(0,2,1000)
q0 = [1,0]

sol = solve_rk4(LCR, t, q0)

plt.figure(figsize=(12,4))

plt.subplot(121)
plt.plot(t, sol[:,0],label='charge(q)')
#plt.plot(t, sol[:,1],label='current(I)')
plt.title('Time trajectory diagram')
plt.xlabel('t')
plt.ylabel('q')
plt.legend()
plt.grid(True)

plt.subplot(122)
plt.plot(sol[:,0], sol[:,1])
plt.title('Phase space diagram')
plt.xlabel('charge(q)')
plt.ylabel('current(I)')
plt.grid(True)

plt.show()
```

```python
# (h) Van der Pol Oscillator equation:
#     x"(t) + mu(1-x^2)x'(t) + x(t) = 0 where mu = 1/n and n = 1,2,3,......, x(0)=1, x'(0)=0
# Let x' = x1 so that x1'(t) = -x - mu(1 - x^2)x1 

from ode_casilab import solve_rk4
import numpy as np 
import matplotlib.pyplot as plt 

def van_der_pol(t, x):
    x, x1 = x
    dx_dt = x1 
    dx1_dt = -x - mu*(1 - x**2)*x1 
    return np.array([dx_dt, dx1_dt])

t = np.linspace(0,10,1000)
mu = 1
x0 = [1,0]

sol = solve_rk4(van_der_pol, t, x0)

plt.figure(figsize=(12,4))

plt.subplot(121)
plt.plot(t, sol[:,0],label='x')
#plt.plot(t, sol[:,1],label='dx_dt')
plt.title('Time trajectory diagram')
plt.xlabel('t')
plt.ylabel('x')
plt.legend()
plt.grid(True)

plt.subplot(122)
plt.plot(sol[:,0], sol[:,1])
plt.title('Phase space diagram')
plt.xlabel('x')
plt.ylabel('dx_dt')
plt.grid(True)

plt.show()
```

```python
# (j) Duffing equation: x"(t) + kx'(t) - x(t) + x^2(t) = -gamma cos(wt)
#     where k = 0.3, gamma = 0.5, w = 1.2, x(0) = x'(0) = 0 
#     let x' = x1 so that x1' + 0.3x1 - x - x^2 = -0.5 cos(1.2t)

from ode_casilab import solve_rk4
import numpy as np 
import matplotlib.pyplot as plt 

def duffing(t, x):
    x, x1 = x
    dx_dt = x1 
    dx1_dt = -0.3*x1 + x + x**2 - 0.5*np.cos(1.2*t) 
    return np.array([dx_dt, dx1_dt])

t = np.linspace(0,10,1000)

x0 = [0,0]

sol = solve_rk4(duffing, t, x0)

plt.figure(figsize=(12,4))

plt.subplot(121)
plt.plot(t, sol[:,0],label='x')
#plt.plot(t, sol[:,1],label='dx_dt')
plt.title('Time trajectory diagram')
plt.xlabel('t')
plt.ylabel('x')
plt.legend()
plt.grid(True)

plt.subplot(122)
plt.plot(sol[:,0], sol[:,1])
plt.title('Phase space diagram')
plt.xlabel('x')
plt.ylabel('dx_dt')
plt.grid(True)

plt.show()
```

```python
# (a) Nonlinear Harmonic oscillator: (theta)"(t) + w^2 sin(theta) = 0, theta(0)=1, theta'(0) = 0
# Let theta' = theta1 so that theta1' = -w^2 sin(theta)

from ode_casilab import solve_rk4
import numpy as np 
import matplotlib.pyplot as plt 

def nonlinear_oscillator(t, theta):
    theta, theta1 = theta
    w = 2
    dtheta_dt = theta1 
    dtheta1_dt = -w**2 *np.sin(theta)
    return np.array([dtheta_dt, dtheta1_dt])

t = np.linspace(0,10,1000)
theta0 = [1, 0]

sol = solve_rk4(nonlinear_oscillator, t, theta0)

plt.plot(t, sol[:,0],label='y')
#plt.plot(t, sol[:,1],label='dy_dt')
plt.title('Time trajectory diagram')
plt.xlabel('t')
plt.ylabel('theta')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# (b) Damped SHO y" + 2(gamma)y' + w^2sin(wy) = 0 
# Let y' = y1 so that y1' + 2(gamma)y1 + w^2sin(wy) = 0
# (i) Over damped system (gamma > w) 
# (ii) Under damped syatem (gamma < w) 
# (iii) Critically damped system (gamma = w)

from ode_casilab import solve_rk4
import numpy as np 
import matplotlib.pyplot as plt 

def SHO(t, y):
    y, y1 = y
    w = 1
    dy_dt = y1 
    dy1_dt = -2*gamma*y1 - w**2 *np.sin(w*y)
    return np.array([dy_dt, dy1_dt])

t = np.linspace(0,10,1000)
y0 = [1,0]

plt.figure(figsize=(7,16))

# (i) Over damped system (gamma > w) 
plt.subplot(311)
gamma = 2
sol = solve_rk4(SHO, t, y0)
plt.plot(t, sol[:,0],label='y')
#plt.plot(t, sol[:,1],label='dy_dt')
plt.title('Over damped system (gamma > w)')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# (ii) Under damped syatem (gamma < w) 
plt.subplot(312)
gamma = .5
sol = solve_rk4(SHO, t, y0)
plt.plot(t, sol[:,0],label='y')
#plt.plot(t, sol[:,1],label='dy_dt')
plt.title('Under damped syatem (gamma < w) ')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# (iii) Critically damped system (gamma = w)
plt.subplot(313)
gamma = 1
sol = solve_rk4(SHO, t, y0)
plt.plot(t, sol[:,0],label='y')
#plt.plot(t, sol[:,1],label='dy_dt')
plt.title('Critically damped system (gamma = w)')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()

plt.grid(True)
plt.show()
```
