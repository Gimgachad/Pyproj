```python
# (b) x' + (v/V)x = m/V , m=1, V=1, x(0)=1
   # => x' + vx = 1, x(0)=1

from ode_casilab import solve_rk4
import numpy as np
import matplotlib.pyplot as plt

def ode_4_b(v,x):
    dx_dv = 1 - v*x
    return dx_dv

v = np.linspace(0,10,1000)
x0 = [1]

sol = solve_rk4(ode_4_b, v, x0)
#print(sol)
plt.plot(v,sol)
plt.xlabel('v')
plt.ylabel('x')
plt.show()
```

```python
# Solution of ODE by Sympy
# from __future__ import division
# x' + vx = 1, x(0)=1
import sympy as sp
x0, v = sp.symbols('x0 v')
x = sp.symbols('x',cls = sp.Function)
eq = sp.Eq(x(v).diff(v,1) + x(v)*v - 1,0)
sol = sp.dsolve(eq, ics={x(v).subs(v,0): 1})
y = sol.simplify()
display(eq)
display(y)
```

```python
# (h) x" + 2t^2x' + x = 0, x(0)=1, x'(0)=0
#    Let x1 = x' so the equation becomes x1' + 2t^2x1 + x = 0

from ode_casilab import solve_rk4
import numpy as np 
import matplotlib.pyplot as plt 

def ode_4_h(t, x):
    x, x1 = x # x1 = x'
    dx_dt = x1
    dx1_dt = -2 * t**2 * x1 - x
    return np.array([dx_dt, dx1_dt])

t = np.linspace(0,10,1000)
x0 = [1,0]

sol = solve_rk4(ode_4_h, t, x0)

plt.plot(t, sol[:,0], label = 'x(t)')
plt.plot(t, sol[:,1], label = 'dx_dt')
plt.xlabel('t')
plt.ylabel('x')
plt.legend()
plt.show()
```

```python
# Solution of ODE by Sympy
# from __future__ import division
# x" + 2t^2x' + x = 0, x(0)=1, x'(0)=0
import sympy as sp
t = sp.symbols('t')
x = sp.symbols('x',cls = sp.Function)
eq = sp.Eq(x(t).diff(t,2) + 2*t**2 *x(t).diff(t,1) + x(t),0)
sol = sp.dsolve(eq, ics={x(t).subs(t,0):1,(x(t).diff(t,1)).subs(t,0):0})
#y = sol.simplify()
display(eq)
display(sol)
```
