```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# parameters
m = 1.0  # mass
k = 1.0  # spring constant
b = 0.2  # damping coefficient
F0 = 10  # amplitude of external force
omega = 1.5  # frequency of external force

# system of differential equations
def system(y, t):
    x, v = y
    dx = v
    dv = (F0*np.sin(omega*t)*np.sin(5*t) - b*v - k*x) / m
    return [dx, dv]

# initial conditions
x0 = 0.0  # initial position
v0 = 0.0  # initial velocity
y0 = [x0, v0]

# time grid
t = np.arange(0, 80, 0.0001)

# solve the ODE
sol = odeint(system, y0, t)

# plot the results
plt.figure(figsize=(14,6))
plt.plot(t, sol[:, 0])
plt.grid(True)
plt.title("Forced mass-spring system")
plt.xlabel('Time')
plt.ylabel('Displacement')
plt.show()
```
