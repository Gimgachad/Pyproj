### 1. Solve the Radioactive Decay equation numerically by python programing using RK-1, RK2, R-3 method and compare all the plots in a single plot also.

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
N0 = 1000  # initial number of atoms
lambda_ = 0.01  # decay constant
t_end = 500  # end time
h = 0.0001  # time step

# Time array
t = np.arange(0, t_end, h)

# Function for the decay equation
def f(N):
    return -lambda_ * N

# RK1 (Euler's method)
N1 = np.zeros_like(t)
N1[0] = N0
for i in range(1, len(t)):
    N1[i] = N1[i-1] + h * f(N1[i-1])

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t, N1, label='RK1 (Euler)')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Number of Atoms')
plt.title('Radioactive Decay using RK1 (Euler\'s Method)')
plt.grid(True)
plt.show()
```

### 2. Solve the Radioactive Decay equation by Sympy module of python and compare it with above four plots. Draw conclusion which of the plots is the best approximation.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from sympy import Function, dsolve, Eq, Derivative, symbols, lambdify

# Define the decay equation
N0, , t_sym = sp.symbols('N0  t')
N = sp.symbols('N', cls=sp.Function)
decay_eq = Eq(Derivative(N(t_sym), t_sym), - * N(t_sym))
print("The differential equation is given by:")
display(decay_eq)

# Solve the decay equation
sol = dsolve(decay_eq)
print("\nand its solution is given by:")
display(sol.simplify())

# Constants
N0_value = 1000  # initial number of atoms
_value = 0.01  # decay constant
t_end = 500  # end time

# Time array
t_values = np.linspace(0, t_end, 1000)

# Substitute the initial condition and solve for the constant
sol = sol.subs('C1', N0_value)
display(sol)

# Convert the solution to a numerical function
N_func = lambdify(t_sym, sol.rhs.subs({N0: N0_value, : _value}), 'numpy')

# Calculate the number of atoms
N_sympy = N_func(t_values)

# Plotting
print("\nAnd the graph is given by:")
plt.figure(figsize=(10, 6))
plt.plot(t_values, N_sympy)
plt.xlabel('Time')
plt.ylabel('Number of Atoms')
plt.title('Radioactive Decay using Sympy')
plt.grid(True)
plt.show()
```
