```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Parameters
m, k, t = sp.symbols('m k t')
b_light = sp.S(0.2) * sp.sqrt(m * k)
b_critical = sp.S(1.0) * sp.sqrt(m * k)
b_heavy = sp.S(2.0) * sp.sqrt(m * k)
b_none = sp.S(0.0)

# Initial conditions
x0 = sp.S(0.0)
v0 = sp.S(3.0)

# Symbolic function
x = sp.Function('x')(t)

def solve_oscillator(b):
    eq = m * x.diff(t, t) + b * x.diff(t) + k * x
    sol = sp.dsolve(eq, x, ics={x.subs(t, 0): x0, x.diff(t).subs(t, 0): v0})
    return sol.rhs

# Solve the differential equations for each case
x_light = solve_oscillator(b_light)
x_critical = solve_oscillator(b_critical)
x_heavy = solve_oscillator(b_heavy)
x_none = solve_oscillator(b_none)

# Substitute numerical values for m and k
m_val, k_val = 1.0, 1.0
solutions = [x_light, x_critical, x_heavy, x_none]
solutions = [sol.subs({m: m_val, k: k_val}) for sol in solutions]

# Create lambdified functions
t_range = np.linspace(0, 50, 1000)
funcs = [sp.lambdify(t, sol, modules=['numpy']) for sol in solutions]

# Evaluate solutions
results = [func(t_range) for func in funcs]

# Plot the results
plt.figure(figsize=(14, 6))
labels = ['Lightly Damped', 'Critically Damped', 'Heavily Damped', 'No Damping']
for res, label in zip(results, labels):
    plt.plot(t_range, res, label=label)

plt.grid(True)
plt.title("Damped and Undamped Mass-Spring System (SymPy Solutions)")
plt.xlabel('Time')
plt.ylabel('Displacement')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Parameters
m = 1.0  # mass
k = 1.0  # spring constant
b = [0.2, 1.0, 2.0]  # damping coefficients for lightly, critically, and heavily damped cases
T = [60, 30, 20]  # total time for each case
N = 1000  # number of time steps
dt = [T[i] / (N - 1) for i in range(3)]  # time steps for each case
t = [np.linspace(0, T[i], N) for i in range(3)]  # time grids for each case

# Initial conditions
x0 = 0.0  # initial position
v0 = 3.0  # initial velocity

# Numerical solution arrays
x_num = [np.zeros(N) for _ in range(3)]

# Initial conditions for numerical solution
for i in range(3):
    x_num[i][0] = x0
    x_num[i][1] = x0 + v0 * dt[i]  # using initial velocity to calculate the second position for 3 cases

# Central finite difference method for damped harmonic oscillator
for j in range(3):
    for i in range(1, N - 1):
        x_num[j][i+1] = (2 * x_num[j][i] - x_num[j][i-1] + (dt[j]**2 / m) * (-b[j] * (x_num[j][i] - x_num[j][i-1]) / dt[j] - k * x_num[j][i]))

# Solve the differential equation using sympy
t_sym = sp.symbols('t')
x_sym = sp.Function('x')(t_sym)
solutions = []

for damping in b:
    eq = m * x_sym.diff(t_sym, t_sym) + damping * x_sym.diff(t_sym) + k * x_sym
    sol = sp.dsolve(eq, x_sym, ics={x_sym.subs(t_sym, 0): x0, x_sym.diff(t_sym).subs(t_sym, 0): v0})
    solutions.append(sol)
    display(sol)

# Convert sympy solutions to numerical functions
def sympy_to_numpy(sol):
    f = sp.lambdify(t_sym, sol.rhs, 'numpy')
    return lambda t: f(t)

x_analytical = [sympy_to_numpy(sol) for sol in solutions]



# Titles for the plots
titles = ["Lightly Damped Mass-Spring System", "Critically Damped Mass-Spring System", "Heavily Damped Mass-Spring System"]

# Plot the solutions
for i in range(3):
    plt.figure(figsize=(14, 6))

    # Numerical solution
    plt.plot(t[i], x_num[i], label="Numerical Solution (CFDM)")

    # Analytical solution
    x_analytical_values = x_analytical[i](t[i])
    plt.plot(t[i], x_analytical_values, label="Analytical Solution (sympy)", linestyle='--')
    
    plt.grid(True)
    plt.title(titles[i])
    plt.xlabel('Time')
    plt.ylabel('Displacement')
    plt.legend()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
m = 1  # mass
k = 1  # spring constant
gamma = [0.2, 1.0, 2.0]  # damping coefficients for underdamped, critically damped, and overdamped cases
T = [60, 30, 20]  # total time for each case
x0 = 0  # initial position
v0 = 3  # initial velocity
t = [np.linspace(0, T[i], 1000) for i in range(3)]  # time range

# Define the angular frequency
omega_0 = np.sqrt(k / m)

# Define the analytical solution for each case
def anal_dhm(t, gamma):
    omega_d = np.sqrt(np.abs(omega_0**2 - gamma**2))  # damped frequency
    if gamma < omega_0:  # Underdamped case
        return np.exp(-gamma * t) * (x0 * np.cos(omega_d * t) + (v0 + gamma * x0) / omega_d * np.sin(omega_d * t))
    elif gamma == omega_0:  # Critically damped case
        return np.exp(-gamma * t) * (x0 + (v0 + gamma * x0) * t)
    else:  # Overdamped case
        r1 = -gamma + omega_d
        r2 = -gamma - omega_d
        C1 = (v0 - r2 * x0) / (r1 - r2)
        C2 = x0 - C1
        return C1 * np.exp(r1 * t) + C2 * np.exp(r2 * t)

# Titles for the plots
titles = ["Underdamped Mass-Spring System", "Critically Damped Mass-Spring System", "Overdamped Mass-Spring System"]

# Plot the solutions
for i in range(3):
    plt.figure(figsize=(14, 8))
    plt.plot(t[i], anal_dhm(t[i], gamma[i]))
    plt.grid(True)
    plt.title(titles[i])
    plt.xlabel('Time')
    plt.ylabel('Displacement')
    plt.show()
```
