```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# Define the system parameters
m = 1.0  # mass
k = 1.0  # spring constant
b = 0.5  # damping coefficient

# Define the system using state space method
A = [[0, 1], [-k/m, -b/m]]  # state matrix
B = [[0], [1/m]]  # input matrix
C = [[1, 0]]  # output matrix
D = [[0]]  # feedforward matrix

# Create a state space system
sys = signal.StateSpace(A, B, C, D)

# Define the time array
t = np.linspace(0, 10, 1000)

# Define the initial conditions (x0 = 1, v0 = 0)
x0 = [1, 0]

# Solve the system
t, y, x = signal.lsim(sys, U=0, T=t, X0=x0)

# Plot the results
plt.figure(figsize=(14,6))
plt.figure()
plt.plot(t, x[:, 0])
plt.title('Displacement vs Time for a Damped Mass-Spring System')
plt.xlabel('Time')
plt.ylabel('Displacement')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.signal import lsim, lti
import matplotlib.pyplot as plt

# Define the circuit parameters
R = 25.0  # Resistance in ohms
C = 0.01  # Capacitance in farads

# Define the system
A = [[0, 1], [-1/(R*C), -1/R]]  # State matrix
B = [[0], [1/C]]  # Input matrix
C = [1, 0]  # Output matrix
D = 0  # Direct feedthrough matrix
system = lti(A, B, C, D)

# Time points
t = np.arange(0, 50, 0.01)

# Initial conditions: q(0) = 1.2*10**(-5)
u = np.zeros_like(t)  # Input signal (0 for all t)
u0 = [1.2*10**(-5), 0]  # Initial state

# Solve the system
t, y, x = lsim(system, u, t, X0=u0)

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(t, x[:, 0], label='Charge (Coulombs)')
plt.legend()
plt.xlabel('Time (Seconds)')
plt.ylabel('Charge')
plt.title('State-Space Solution for an LCR Circuit Discharging')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.signal import lsim, TransferFunction
import matplotlib.pyplot as plt

# Define the circuit parameters
R = 1.0  # Resistance in ohms
L = 1.0  # Inductance in henries
C = 1.0  # Capacitance in farads

# Define the system
num = [1/L]  # Numerator of the transfer function
den = [1, R/L, 1/(L*C)]  # Denominator of the transfer function
system = TransferFunction(num, den)

# Time points
t = np.linspace(0, 10, 1000)

# Initial conditions: x1(0) = 1, x2(0) = 0
u = np.zeros_like(t)  # Input signal (0 for all t)
u0 = [1/C, 0]  # Initial state

# Solve the system
t, y, x = lsim(system, u, t, X0=u0)

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(t, x[:, 0], label='Charge (Coulombs)')
plt.plot(t, x[:, 1], label='Current (Amperes)')
plt.legend()
plt.xlabel('Time (Seconds)')
plt.ylabel('State Variables')
plt.title('State-Space Solution for an LCR Circuit Discharging')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# Parameters for a forced mass-spring system
m = 1.0  # mass
k = 1.0  # spring constant
b = 0.5  # damping coefficient
F = 1.0  # force amplitude

# State-space matrices
A = [[0, 1], [-k/m, -b/m]]
B = [[0], [1/m]]
C = [[1, 0]]
D = [[0]]

# Create a state-space system
sys = signal.StateSpace(A, B, C, D)

# Time array
t = np.linspace(0, 10, 1000)

# Input array (force)
u = F * np.sin(t)

# Solve the system
t, y, x = signal.lsim(sys, u, t)

# Plot the results
plt.figure()
plt.plot(t, y)
plt.title('Forced Mass-Spring System Response')
plt.xlabel('Time [s]')
plt.ylabel('Displacement [m]')
plt.grid(True)
plt.show()
```
