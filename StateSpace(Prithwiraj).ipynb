```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# Define the system parameters
m = 1.0  # mass
k = 1.0  # spring constant
b = 0.5  # damping coefficient

# Define the system using state space method
A = [[0, 1], [-k/m, -b/m]]  # state matrix
B = [[0], [1/m]]  # input matrix
C = [[1, 0]]  # output matrix
D = [[0]]  # feedforward matrix

# Create a state space system
sys = signal.StateSpace(A, B, C, D)

# Define the time array
t = np.linspace(0, 10, 1000)

# Define the initial conditions (x0 = 1, v0 = 0)
x0 = [1, 0]

# Solve the system
t, y, x = signal.lsim(sys, U=0, T=t, X0=x0)

# Plot the results
plt.figure(figsize=(14,6))
plt.figure()
plt.plot(t, x[:, 0])
plt.title('Displacement vs Time for a Damped Mass-Spring System')
plt.xlabel('Time')
plt.ylabel('Displacement')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# Parameters for an LCR circuit
L = 1.0  # inductance
C = 1.0  # capacitance
R = 1.0  # resistance
V = 1.0  # voltage

# State-space matrices
A = [[0, 1], [-1/(L*C), -R/L]]
B = [[0], [1/L]]
C = [[1, 0]]
D = [[0]]

# Create a state-space system
sys = signal.StateSpace(A, B, C, D)

# Time array
t = np.linspace(0, 10, 1000)

# Input array (voltage source)
u = V * np.ones_like(t)

# Solve the system
t, y, x = signal.lsim(sys, u, t)

# Plot the results
plt.figure()
plt.plot(t, y)
plt.title('LCR Circuit Response')
plt.xlabel('Time [s]')
plt.ylabel('Current [A]')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# Parameters for a forced mass-spring system
m = 1.0  # mass
k = 1.0  # spring constant
b = 0.5  # damping coefficient
F = 1.0  # force amplitude

# State-space matrices
A = [[0, 1], [-k/m, -b/m]]
B = [[0], [1/m]]
C = [[1, 0]]
D = [[0]]

# Create a state-space system
sys = signal.StateSpace(A, B, C, D)

# Time array
t = np.linspace(0, 10, 1000)

# Input array (force)
u = F * np.sin(t)

# Solve the system
t, y, x = signal.lsim(sys, u, t)

# Plot the results
plt.figure()
plt.plot(t, y)
plt.title('Forced Mass-Spring System Response')
plt.xlabel('Time [s]')
plt.ylabel('Displacement [m]')
plt.grid(True)
plt.show()
```
