### 0.1.4 Recurssive function:

```python
def factorial(n):
    if n == 0:
        return 1

    else:

        return n * factorial(n-1)
    
# Example usage
print(factorial(5))
print(factorial(8))    
```

```python
# Input based Recursive Function
n = int(input("Enter an integer: "))

def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

# Example usage
print (fibonacci (n))
```

### 0.1.5 Plot from Recursive function:

```python
# Recursive Plot of a function
import numpy as np
import matplotlib.pyplot as plt

def sine_function(x):
    y = np.sin(x)
    return y

x_axis = np.arange(0, 4*np.pi, 0.1)
y_axis = sine_function(x_axis)

# Plotting of wave function
plt.plot(x_axis, y_axis, label=r"$x$ versus $y$ Plot")
plt.grid(color="g", linestyle = "--", linewidth = .5)
plt.title("My Plot")
plt.ylabel("Sin(x)")
plt.xlabel("x")
plt.legend(loc="best")
plt.show()
```

```python
# Recurssive plot with loop inside a function
import numpy as np
import matplotlib.pyplot as plt

# Recursive Define an empty set y=[] and fill it up recursively
x = np.arange(0, 2*np.pi, 0.1) # x_axis is a Global function

# This function will generate y value
def sine_function(x):
    y = []
    for val in x:
        y.append (np. sin(val))
    return y

# Generate y values using sine_function
y = sine_function(x)

# Plotting of wave function
plt.plot(x, y, label=r"$x$ versus $y$ Plot")
plt.grid(color="g", linestyle="--", linewidth=.5)
```

```python
def linear_interpolation(x, y, xp):
# Find the index of the largest r value less than or equal to xp
    i = []
    for idx, val in enumerate(x):
        if val <= xp:
            i = idx

# Perform linear interpolation
    yp = y[i] + (y[i+1] - y[i]) * (xp - x[i]) / (x[i+1] - x[i])

    return yp

# Eample usage
time = [0, 1, 2, 3, 4, 5] # Example list of time values
temperature = [20, 25, 30, 35, 40, 45] # Example list of temperature values

# Interpolate the temperature at time 2.5
interpolated_temp = linear_interpolation(time, temperature, 2.5)
print ("Interpolated temperature at time 2.5:", interpolated_temp)

```

```python
# Linear Interpolation_new
def linear_interpolation(x, y, xp):
    # Find the inder of the largest x value less than or equal to xp
    i = max(idx for idx, val in enumerate(x) if val <= xp)
    # Perform linear interpolation
    yp = y[i] + (y[i+1] - y[i]) * (xp - x[i]) / (x[i+1] - x[i])
    return yp

# Example usage
time = [0, 1, 2, 3, 4, 5] # Example list of time values
temperature = [20, 25, 30, 35, 40, 45] # Example list of temperature values

# Interpolate the temperature at time 2.5
interpolated_temp = linear_interpolation(time, temperature, 2.5)
print("Interpolated temperature at time 2.5 is:", interpolated_temp)

```

```python
# x axis

# Example I: Plotting using Lambda function
import numpy as np
import matplotlib.pyplot as plt

# Global parameters
A = B = 0.5
omega = 2 * np.pi

# Define lambda functions for x and y
lambda_x = lambda t: A * np.sin(omega * t)
lambda_y = lambda t: B * np.cos(omega * t)

# Generate x values using lambda_t
lambda_t = np.arange(0, 2 * np.pi, 0.00001)

# Calculate y values using lambda_r and lambda_y
lambda_x_values = lambda_x(lambda_t)
lambda_y_values = lambda_y(lambda_t)

# Plotting of wave function
plt.plot (lambda_x_values, lambda_y_values, label=r"$x(t)$ versus $y(t) $ Plot")
plt.grid(color="g", linestyle="--", linewidth=0.5)
plt.title("Parametric Plot of (x(t),y(t))")
plt.xlabel("x(t)")
plt.ylabel("y(t)")
plt. legend(loc="best")
plt. show ()
```

```python
# Example II: Plotting using Lambda function
import numpy as np
import matplotlib.pyplot as plt

# Global parameters
A = [0.5, 1.0, 1.5, 2.0]
omega = 2 * np.pi

# Define lambda functions for x and y
lambda_x = lambda t, a: a * np.sin(omega * t)
lambda_y = lambda t, a: a * np. cos(omega * t)

# Generate x values using lambda_t
lambda_t = np.arange(0, 2 * np.pi, 0.00001) # x_axis

# Calculate x and y values for each value of A
lambda_x_values = [lambda_x(lambda_t, a) for a in A]
lambda_y_values = [lambda_y(lambda_t, a) for a in A]

# Plotting of wave function for each value of A
for i, a in enumerate(A):
    plt.plot (lambda_x_values [i], lambda_y_values[i], label=f"A={a}")

plt.grid(color="g", linestyle="--", linewidth=0.5)
plt.title("Parametric Plot of (x(t),y(t))")
plt.xlabel("x(t)")
plt.ylabel("y(t)")
plt.legend(loc="best")
plt. show ()
```

```python
# Example II: Plotting using Lambda function
import numpy as np
import matplotlib.pyplot as plt

# Global parameters
A = [0.5, 1.0, 1.5, 2.0]
omega = 2 * np.pi

# Generate x values using lambda_t
lambda_t = np.arange(0, 2 * np.pi, 0.00001) # r_axis

# Plotting of wave function for each value of A
for a in A:
# Define lambda functions for x and y inside the loop
    lambda_x = lambda t: a * np.sin(omega * t)
    lambda_y = lambda t: a * np.cos(omega * t)

    # Calculate x and y values for the current value of A
    lambda_x_values = lambda_x(lambda_t)
    lambda_y_values = lambda_y(lambda_t)

    # Plot the parametric curve (x(t), y(t)) for the current value of A
    plt.plot(lambda_x_values, lambda_y_values, label=f"A={a}")

plt.grid(color="g", linestyle="--", linewidth=0.5)
plt.title("Parametric Plot of (x(t),y(t))")
plt.xlabel("x(t)")
plt.ylabel("y(t)")
plt.legend(loc="best")
plt. show ()
```
