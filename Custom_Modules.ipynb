```python
import numpy as np
import matplotlib.pyplot as plt
from priyangshu import rk2

# Define the differential equation dy/dx = xy
def differential_equation(y, x):
    dydx = x * y
    return dydx

# Set up time points (or x-values in this case)
x = np.linspace(0, 2, 10000)

# Set initial conditions for the differential equation
y0 = 5.0

# Use RK2 solver to solve the differential equation
solution = rk2(differential_equation, y0, x)

# Plot the solution
plt.plot(x, solution, label='Solution for dy/dx = x * y')
plt.title('Solution of dy/dx = x * y')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from priyangshu import rk4

# Define Lorenz attractor equations
def lorenz_system(y, t):
    sigma = 10.0
    rho = 28.0
    beta = 8.0 / 3.0
    x, y, z = y
    dxdt = sigma * (y - x)
    dydt = rho * x - y - x * z
    dzdt = x * y - beta * z
    return np.array([dxdt, dydt, dzdt])

# Set up time points
t = np.linspace(0, 50, 10000)

# Set initial conditions for Lorenz attractor
y0_lorenz = np.array([1.0, 0.0, 0.0])

# Solve Lorenz attractor using RK4
solution_lorenz = rk4(lorenz_system, y0_lorenz, t)

# Plot Lorenz attractor
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(solution_lorenz[:, 0], solution_lorenz[:, 1], solution_lorenz[:, 2], lw=0.5)
ax.set_title('Lorenz Attractor')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from priyangshu import rk4

# Lotka-Volterra equations
def lotka_volterra(y, t, alpha, beta, gamma, delta):
    x, z = y
    dxdt = alpha * x - beta * x * z
    dydt = delta * x * z - gamma * z
    return np.array([dxdt, dydt])

# Define a function with the Lotka-Volterra parameters
def lotka_volterra_with_params(y, t):
    alpha, beta, gamma, delta = 1.1, 0.4, 0.4, 0.1
    return lotka_volterra(y, t, alpha, beta, gamma, delta)

# Set up time points
t = np.linspace(0, 20, 1000)

# Set initial conditions for the Lotka-Volterra equations
initial_conditions = np.array([2, 1])  # Initial prey and predator populations

# Use RK4 solver to solve the Lotka-Volterra equations
solution_lv = rk4(lotka_volterra_with_params, initial_conditions, t)

# Plot the solutions
plt.figure(figsize=(10, 6))
plt.plot(t, solution_lv[:, 0], label='Prey (x)')
plt.plot(t, solution_lv[:, 1], label='Predator (y)')
plt.title('Lotka-Volterra Equations with RK4')
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.show()
```
