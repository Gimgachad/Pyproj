```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply by kind and color')
ax.legend(title='Fruit color')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-2*np.pi,2*np.pi,900)
y=np.tan(x)
# This operation inserts a NaN where the difference between successive points is negative
y[:-1][np.diff(y) < 0] = np.nan
# Mask values where absolute value of y is greater than a large number
y = np.ma.array(y).filled(np.nan)
y[np.abs(y) > 20] = np.nan

# Create a new figure and a subplot
fig, ax = plt.subplots()

plt.plot(x,y)
plt.title("tan wave")
plt.xlabel("time")
plt.ylabel("function")
plt.grid(True)
plt.show()
```
