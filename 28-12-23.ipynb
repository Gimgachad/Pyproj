```python
#30 to plot a equation using numpy and matplotlib
import numpy as np, matplotlib.pyplot as plt
x = np.linspace(1,10,1000)
linear_line = 0.5*x+4
log_curve = np.log(x)
sine_wave = np.sin(x)
curves = [linear_line, log_curve, sine_wave]
colors = ['k','r','b']
styles = ['-','--',':']
for curve, color, style in zip(curves,colors,styles):
    plt.plot(x, curve, c=color, linestyle=style)
plt.title("Three plots")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#34 LIFTOFF! Program with yield(instead of 'return')
def countdown(n):
    while n>0:
        yield n
        n-=1
    yield "LIFTOFF!"
for i in countdown(4):
    print(i)
```
