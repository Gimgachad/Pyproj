```python
import numpy as np
import matplotlib.pyplot as plt

# Read data from CSV file using NumPy
data = np.genfromtxt("D:/Downloads/data.csv", delimiter = ',')

# Extract x and y values from the data
x = data[:, 0] # Assuming x values are in the first column
y = data[:, 1] # Assuming y values are in the second column

# Perform linear regression to find the best fit line
slope, intercept = np.polyfit(x[1:], y[1:], 1)

# Calculate the best fit line
best_fit_line = slope * x + intercept

# Plot the data and the best fit lline
plt.scatter(x, y, label='Data')
plt.plot(x, best_fit_line, color='red', label='Best Fit Line')

# Add labels and title
plt.xlabel('x')
plt.ylabel('Y')
plt.title('Best Fit Line from CSV Data')

# Add legend
plt.legend()
plt.grid(True)
plt.show()
```
