### Ex-1: Find the numerical solution of the following differential eqn over the domain [0,2]:

$y\prime = xy,              y(0) =1$

```python

# The numerical solution can be programmed by a single bor-loop ob x with increment ob
# h. In the beginning, debine the given equation in an inline function and suppose h = 0.5.
def f(x,y):
    return x * y

x = 0 # intitial value of x
xn = 2 # final value of x
y = 1 # value of y(x0)
h = 0.5 # step-size
n = int((xn - x) / h)
print('x \t\t y (Euler)') # Data table header
print('%f \t %f' % (x,y)) # tabular format output
for i in range(n):
    y += f(x,y) * h # calculate next y
    x += h  # increment x
    print('%f \t %f' % (x,y))

# Since the initial values are known, the loop starts form the next x value, and the derivative in
# this example is function of y only.
```

### By using RK4, find the numerical solution of the following differential equation for the domain [2, 4] and plot the both solutions. Use step size h = 0.1:

$y\prime=y+3, y(2)=-2$

```python
import numpy as np
import matplotlib.pyplot as plt

def b(x, y):
    return y + 3

def a(x):
    return np.exp(x - 2) - 3

x = 2
xn = 4
y = -2
h = 0.1
n = int((xn - x) / h)
xp = np.linspace(x, xn, n+1) # x array for plot
yp = np.empty(n+1, float)
yp[0] = y
print('x \t\t y (RK4) \t y (analytical)')
print('%f \t %f \t %f' % (x,y,a(x)))
for i in range (1, n+1):
    K1 = b(x, y) * h 
    K2 = h * b(x + h / 2, y + K1 / 2)
    K3 = h * b(x + h / 2, y + K2 / 2)
    K4 = h * b(x + h, y + K3)
    y += (K1 + 2 * K2 + 2 * K3 + K4)
    yp[i] = y
    x += h  # increment x
    for xi in xp:
        print('%f \t %f \t %f' % (x, y, a(xi)))


plt.plot(xp, yp, 'ro' )
plt.plot(xp, a(xp))
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['RK4', 'Analytical'])
plt.show()
```

### Find the solution of the 2nd order ODE over domkain [$\pi$,$2\pi$]:

$y\prime\prime+y=4x+10\sin(x), y(\pi)=0, y\prime(\pi)=2$

```python
import numpy as np
import matplotlib.pyplot as plt

def dy(x, y, u):
    return u

def du(x, y , u):
    return 4 * x + 10 * np.sin(x) - y

#analytical
def anal(x):
    return 9 * np.pi * np.cos(x) + 7 * np.sin(x) + 4 * x - 5 * x * np.cos(x)

def danal(x):
    return - 9 * np.pi * np.sin(x) - 7 * np.cos(x) + 4 - 5 * (np.cos(x) - x * np.sin(x))

x = np.pi
xn = 2 * np.pi
y = 0.0
u = 2
h = 0.1
n = int((xn - x) / h)
xp = np.linspace(x, xn, n+1) # x array for plot
yp = np.empty(n+1, float)
yp[0] = y
up = np.empty(n+1, float)
up[0] = u
print('x \t\t y\' (RK4) \t y(RK4) \t y\' (Exact) \t y(Exact)')
print('%f \t %f \t %f \t %f \t %f' % (x, u, y, danal(x), anal(x)))
for i in range (1, n+1):
    L1 = h * du(x, y, u)
    K1 = dy(x, y, u) * h 
    L2 = h * du(x + h / 2, y + K1 / 2, u + L1 / 2)
    K2 = h * dy(x + h / 2, y + K1 / 2, u + L1 / 2)
    L3 = h * du(x + h / 2, y + K2 / 2, u + L2 / 2)
    K3 = h * dy(x + h / 2, y + K2 / 2, u + L2 / 2)
    L4 = h * du(x + h, y + K3, u + L3)
    K4 = h * dy(x + h, y + K3, u + L3)
    u += (L1 + 2 * L2 + 2 * L3 + L4) / 6
    up[i] = u
    y += (K1 + 2 * K2 + 2 * K3 + K4) / 6
    yp[i] = y
    x += h  # increment x
    for xi in xp:
        print('%f \t %f \t %f \t %f \t %f' % (x, u, y, danal(x), anal(x)))

plt.figure(figsize=(14,7))
plt.plot(xp, yp, color = 'r', marker = 'o' , ls = '', label = 'y (RK4)')
plt.plot(xp, up, color = 'b', marker = 's', ls = '' , label = 'y\' (RK4)')
plt.plot(xp , anal(xp) , color = 'k' , lw = 1.5 , ls = '-' , label = 'y (Exact)')
plt.plot(xp , danal(xp) , color = 'k' , lw = 1.5 , ls = '--' , label = 'y\' (Exact)')

plt.xlabel('x')
plt.ylabel('y, y\'')
plt.legend(loc = 'upper left')
plt.grid()
plt.axis([np.pi , 2 * np.pi , None , None])
plt.show()
```
