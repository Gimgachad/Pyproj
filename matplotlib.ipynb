```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

N0=1000
gamma=0.1
t=np.arange(0,100)
Nt=N0*(np.e**(-gamma*t))

# Plotting
print("\nAnd the graph is given by:")
plt.figure(figsize=(10, 6))
plt.plot(t,Nt)
plt.xlabel('Time')
plt.ylabel('Number of Atoms')
plt.grid(True)
plt.show()

```

```python
# Radioactive Decay(RK2 method)
import numpy as np
import matplotlib.pyplot as plt

def radio_active_decay(t,N):
    dN_dt = -gamma*N
    return dN_dt
# Constants
gamma = 0.1
# Parameters
t,N,h = 0,1000,0.01

tlist = []
Nlist = []
for i in range(0,10000):
    k1 = h*radio_active_decay(t,N)
    k2 = h*radio_active_decay(t + h,N + k1)
    t += h
    N += (k1+k2)/2
    tlist.append(t)
    Nlist.append(N)

plt.plot(tlist,Nlist,label='Radioactive Decay(RK2 method)')
plt.xlabel('Time')
plt.ylabel('N(t)')
plt.grid(True)
plt.legend()
plt.show()
```

```python

# Radioactive decay using 4th order R-K method
import numpy as np
import matplotlib.pyplot as plt

def radio_active_decay(t,N):
    dN_dt = -gamma*N
    return dN_dt

gamma = 0.1

t,N,h = 0,1000,0.01

tlist = []
Nlist = []

for i in range(0,10000):
    k1 = h*radio_active_decay(t,N)
    k2 = h*radio_active_decay(t + h/2,N + k1/2)
    k3 = h*radio_active_decay(t + h/2,N + k2/2)
    k4 = h*radio_active_decay(t + h,N + k3)
    
    tlist.append(t)
    Nlist.append(N)
    t += h
    N += (k1+2*(k2+k3)+k4)/6

plt.plot(tlist,Nlist,label='Radioactive Decay(RK4 method)')
plt.xlabel('Time')
plt.ylabel('N(t)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plotting in same plot
plt.figure(figsize=(10, 6))
plt.plot(tlist, Nlist, label='RK2 (Heun)')
plt.plot(tlist, Nlist,'--', label='RK4 (Kutta)')
plt.plot(t,Nt,':',label='Matplotlib')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Number of Atoms')
plt.title('Radioactive Decay using Different RK Methods')
plt.grid(True)
plt.show()
```
