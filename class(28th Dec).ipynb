```python
#27
#COnvert binary to decimal with print
# Function to convert binary to decimal
def binary_to_decimal(binary):
    decimal = 0
    for digit in binary:
        decimal = decimal*2 + int(digit)
    return decimal
binary_number=input("Enter a binary no.:")
print(binary_to_decimal(binary_number))
```

```python
#To plot a equation using 'Numpy' and 'Matplotlib'
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace (1, 10, 1000)
linear_line= 0.5*x+4
log_curve=np.log(x)
sin_wave = np.sin(x)
curves=[linear_line, log_curve, sin_wave]
colors=['k', 'r', 'b']
styles=['-', '--',':']
#for curve, color, style in zip(curves, colors,styles):
for curve, color, style in zip(curves, colors,
styles):
    plt.plot(x, curve, c=color, linestyle=style)
plt.title("Three plots")
plt.xlabel("x-axis title")
plt.ylabel("y-axis title")
plt.show()
```

```python
#LIFTOFF! Program with yield (instead of "return")
def countdown(n):
    while n>0:
        yield n
        n-=1
    yield "LIFTOFF!"
for i in countdown(4):
    print(i)
```
